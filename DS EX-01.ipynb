```python
df['show_name']=df['show_name'].fillna(df['show_name'].mode()[0])
df.head()
```

```python
df['rating']=df['rating'].fillna(df['rating'].mean())
df.head()
```

```python
df['watchers']=df['watchers'].fillna(df['watchers'].median())
df.head()
```
